# Projeto Python IA: Inteligência Artificial e Previsões

### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom



```python
#1- Importar a base de dados
import pandas as pd

tabela = pd.read_csv('clientes.csv')

display(tabela)



```

```python
#verificar se a erros na tabela

display(tabela.info())
```

```python
#2- Preparar a base de dados para a inteligencia artificial
    #Pre processamento
from sklearn.preprocessing import LabelEncoder

#Tranformar colunas de texto em numero(Label encoder):
codificador = LabelEncoder()

#Profissão
tabela['profissao'] = codificador.fit_transform(tabela['profissao'])

#mix de credito
tabela['mix_credito'] = codificador.fit_transform(tabela['mix_credito'])

#comportamento_pagamento
tabela['comportamento_pagamento'] = codificador.fit_transform(tabela['comportamento_pagamento'])

display(tabela)
```

```python
#X = Dados que a IA usa para fazer a previsão
#Y = Dado para ser previsto

x = tabela.drop(columns=['score_credito', 'id_cliente'])

y = tabela['score_credito']

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)
 
```

```python
#3- Criar um modelo de IA -> Nota de credito: ruim, media ou boa
#arvore de decisão - Randomforest:
#neirest neighbors -> KNN:
#importar IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

#criar IA
modelo_arvore = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

#treinar a IA
modelo_arvore.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
#4- Escolher o melhor modelo

previsão_arvore = modelo_arvore.predict(x_teste)
previsão_knn = modelo_knn.predict(x_teste)

from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste, previsão_arvore))
display(accuracy_score(y_teste, previsão_knn))
 
```

```python
#5- Usa o modelo escolhido para fazer novas previsões
#modelo de arvore de decisão é o melhor para este caso

#importar nova base de dados de clientes
tabela_novos_clientes = pd.read_csv('novos_clientes.csv')

#tranformar as colunas de texto da nova base

#Profissão
tabela_novos_clientes['profissao'] = codificador.fit_transform(tabela_novos_clientes['profissao'])

#mix de credito
tabela_novos_clientes['mix_credito'] = codificador.fit_transform(tabela_novos_clientes['mix_credito'])

#comportamento_pagamento
tabela_novos_clientes['comportamento_pagamento'] = codificador.fit_transform(tabela_novos_clientes['comportamento_pagamento'])


#executar nossa IA na nova base de dados
previsoes = modelo_arvore.predict(tabela_novos_clientes)

display(tabela_novos_clientes)
display(previsoes)
```
